```python
# a helper class for imporing the information about amplitude and period of the wells
class Coordinates:
    def __init__(self,name):
        file=open(name)
        file.readline()
        self.shapes = np.array([np.array(list(map(int, line.split(',')))) for line in file])
        file.close()
        print(len(self.shapes))
        print(self.shapes[-1])
        self.shapesraw=self.shapes

        print("removing wells with problems")
        self.shapes=self.shapes[self.shapes[:,4]==0]
        print(len(self.shapes))

        self.T100A20=self.shapes[np.logical_and(self.shapes[:,3]==100,self.shapes[:,2]==20)][:,[0,1]]
        self.T50A20=self.shapes[np.logical_and(self.shapes[:,3]==50,self.shapes[:,2]==20)][:,[0,1]]
        self.T20A20=self.shapes[np.logical_and(self.shapes[:,3]==20,self.shapes[:,2]==20)][:,[0,1]]
        self.T10A20=self.shapes[np.logical_and(self.shapes[:,3]==10,self.shapes[:,2]==20)][:,[0,1]]
        self.T100A10=self.shapes[np.logical_and(self.shapes[:,3]==100,self.shapes[:,2]==10)][:,[0,1]]
        self.T50A10=self.shapes[np.logical_and(self.shapes[:,3]==50,self.shapes[:,2]==10)][:,[0,1]]
        self.T20A10=self.shapes[np.logical_and(self.shapes[:,3]==20,self.shapes[:,2]==10)][:,[0,1]]
        self.T10A10=self.shapes[np.logical_and(self.shapes[:,3]==10,self.shapes[:,2]==10)][:,[0,1]]
        self.flat=self.shapes[self.shapes[:,3]==0][:,[0,1]]
        self.flat_up=self.shapes[self.shapes[:,3]==0][:,[0,1]][0:40]
        self.flat_down=self.shapes[self.shapes[:,3]==0][:,[0,1]][40:]

        self.TAs=[self.flat_up,self.flat_down,self.T100A10,self.T100A20,self.T50A10,self.T50A20,self.T20A10,
               self.T20A20,self.T10A10,self.T10A20]
```

```python
# function for plotting the average of many replicate time series
def plot_av(data,qx=0.5,out_data=False,ax=None,skip=1):
    tmp=np.array(data)
    av=np.mean(tmp[:,::skip,1],axis=0)
    nsam=len(tmp[:,0,0])
    std=np.std(tmp[:,::skip,1],axis=0)/np.sqrt(nsam)
    if (out_data==True):
        return tmp[0,:,0]*qx,av,std
    else:
        if ax==None:
            plt.errorbar(tmp[0,::skip,0]*qx,av,std)
        else:
            ax.errorbar(tmp[0,::skip,0]*qx,av,std)
```

```python
def secdiff(x):
    dx=np.diff(x)
    return dx/(x[-1]-x[0])

def avdiff(x):
    dx=np.diff(x)
    return np.mean(dx)/(x[-1]-x[0])


# a helper class for reading and processing data files
class SectorData:
    def __init__(self, name):
        file=open(name)
        print(file.readline())
        self.secs = [np.array(list(map(int, line.split(',')))) for line in file]
        file.close()
        self.secsizes=[np.concatenate((s[0:4],s[9:])) for s in self.secs]
        self.secs=np.array([s[0:9] for s in self.secs])
        print(self.secs[-1])
        print(self.secsizes[-1])
        self.tend=np.max(self.secs[:,1])
        
    def avsecsize(self,wells):
        #pi,ti,zi,ni,xposs
        tmp=[]
        for s in wells:
            pi,ni=s[0],s[1]
            tmp.append([[xx[1],avdiff(xx[4:])] for xx in self.secsizes if xx[0]==pi and xx[3]==ni])
        return tmp
    def avsecsizeend(self,wells): 
        tmp=[]
        for s in wells:
            pi,ni=s[0],s[1]
            tmp.append([avdiff(xx[4:]) for xx in self.secsizes if xx[0]==pi and xx[3]==ni and xx[1]>self.tend-20])
        return np.mean(tmp), np.std(tmp)/np.sqrt(len(tmp))
    def avsecsizeendlist(self,wells): 
        tmp=[]
        for s in wells:
            pi,ni=s[0],s[1]
            tmp.append([avdiff(xx[4:]) for xx in self.secsizes if xx[0]==pi and xx[3]==ni and xx[1]==self.tend-1])
        return np.array(tmp)
    def avsecR(self,wells):
        tmp=[]
        for s in wells:
            pi=s[0]
            ni=s[1]
            sel=np.logical_and(self.secs[:,0]==pi,self.secs[:,3]==ni)
            if (sel.any()==True): tmp.append([[x,1.*r/(r+s+1e-6)] for x,s,r in self.secs[sel][:,[1,4,5]]])
        return tmp
    def avsecS(self,wells):
        tmp=[]
        for s in wells:
            pi=s[0]
            ni=s[1]
            sel=np.logical_and(self.secs[:,0]==pi,self.secs[:,3]==ni)
            if (sel.any()==True): tmp.append([[x,1.*s/(r+s+1e-6)] for x,s,r in self.secs[sel][:,[1,4,5]]])
        return tmp
    def avsecST(self,ti,wells):
        tmp=[]
        for s in wells:
            pi=s[0]
            ni=s[1]
            sel=np.logical_and(np.logical_and(self.secs[:,0]==pi,self.secs[:,3]==ni),self.secs[:,1]==ti)
            if (sel.any()==True): tmp.append([1.*s/(r+s) for s,r in self.secs[sel][:,[4,5]]])
        return np.concatenate(tmp)
    def avsector(self,ti,wells):
        tmp=self.avsecST(ti,wells)
        return np.mean(tmp), np.std(tmp)/np.sqrt(len(tmp))
```

```python
plt.rcParams["figure.figsize"] = (8,2.6)
plt.rcParams['font.size'] = '10'

plt.figure(facecolor='w')
fig,ax=plt.subplots(1,3,constrained_layout=True)
fig.patch.set_facecolor('w')

TAs=[c1.flat_down,c1.T10A20,c1.T20A20,c1.T50A20,c1.T100A20]
for TA in TAs:
    plot_av(s1.avsecS(TA),ax=ax[0],skip=4)
ax[0].set_xlabel("time [h]"); ax[0].set_ylabel("mean S fraction")
ax[0].set_ylim(bottom=0)
ax[0].vlines((129/2,273/2),0,1.2,colors='black')

TAs=[c2.flat_down,c2.T10A20,c2.T20A20,c2.T50A20,c2.T100A20]
for TA in TAs:
    plot_av(s2.avsecS(TA),ax=ax[1],skip=4)
ax[1].set_xlabel("time [h]"); #plt.ylabel("mean S fraction")
ax[1].set_ylim(bottom=0)
ax[1].vlines((135/2,331/2),0,1.2,colors='black')

TAs=[c3.flat_up,c3.T10A20,c3.T20A20,c3.T50A20,c3.T100A20]
for TA in TAs:
    plot_av(s3.avsecS(TA),ax=ax[2],skip=4)
ax[2].legend(('flat','T10A20','T20A20','T50A20','T100A20'),loc='lower left')
ax[2].set_xlabel("time [h]"); #plt.ylabel("mean S fraction")
ax[2].set_ylim(bottom=0)
ax[2].vlines((243/2,385/2),0,1.2,colors='black')
fig.savefig(currdir+'\\data_Fig_4\\1_step_fractions_time.png',dpi=300)
plt.show()
```
