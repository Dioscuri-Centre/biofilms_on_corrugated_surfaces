# Code for generating plots for Fig. S5BCD

```python
from matplotlib import pyplot as plt
import os
import numpy as np
from scipy.signal import convolve
import matplotlib as mpl
import json
```

```python
# times at which 60s time-lapse videos were taken. t=0 is immediately after inoculation
times_from_t0=[20.4, 29.1, 68.6, 146.2, 166.6, 191.8]
```

```python
currdir=os.getcwd()
```

```python
pixelsize=0.4172 # microns/pixel
```

```python
# positions of the six wells in the original 60-p dataset
origpos=[0, 6, 13, 21, 29, 30]
```

### The code below processes velocity fields obtained using the python script 'velocity_profiles_v2_revision.py' using raw data from  "2023_11_14_velocities\2023_11_14_velocities_1\". 

```python
def bin_xy(data,qx=1,binsize=1):
    # data must be in the form of a list of (x,y) lists of coordinates
    # x,y can have different length for different lists
    nsam=len(data)
    #print("nsam=",nsam)
    x1=np.min([np.min(d[0]) for d in data])
    x2=np.max([np.max(d[0]) for d in data])
    nx=int((x2-x1)/binsize)
    xx=np.linspace(x1,x2,nx+1)
    yy=np.full((nx+1),0.) 
    yy2=np.full((nx+1),0.) 
    nn=np.full((nx+1),0) 
    for d in data:
        x,y=d
        inds=np.array(nx*(np.array(x)-x1)/(x2-x1),dtype=int)
        yy[inds]+=y
        yy2[inds]+=y*y
        nn[inds]+=1
        #print(inds)
    yy=yy/(nn+1e-12)
    yy2=yy2/(nn+1e-12)
    yy=yy[nn>0]
    yy2=yy2[nn>0]
    xx=xx[nn>0]
    nn=nn[nn>0]
    yy2=np.sqrt((yy2-yy*yy)/nn)
    return xx,yy,yy2 # x,y and error

class VelocityField:
    def __init__(self, name,verbose=True,order=[]):
        self.alldata=[]
        
        # first import data
        with open(name, 'rb') as f:
            for i in range(0,100):
                try:
                    coords=np.load(f) # [pi,ni,x1,y1,x2,y2]
                    data=np.load(f)
                    snapshot=np.load(f)
                    snapshot=0.5+0.5*snapshot/np.max(snapshot)
                    self.alldata.append((coords,data,snapshot))
                    if (verbose): print(i,": ",coords)
                except ValueError:
                    if (verbose): print("end of file")
                    break
        # and the process it to get the vertical component of the velocity field for each well
        self.vys=[]
        self.vys_aver=[]
        p2=0
        p3=0
        for p in order:
            for i in range(4*p,4*p+4):  
                coords,data,snapshot=self.alldata[i]
                vy=data[3]
                dx=16
                ny,nx=vy.shape
                y = np.linspace(dx/2.+dx*(ny-1),dx/2.,ny)
                vely=-np.mean(vy[:,4:(len(vy[0])-4)],axis=1)*pixelsize
                if (p!=order[-1]): # wells in the upper channel
                    ybottom=0
                    dist=(y[0]-y-ybottom)*pixelsize
                    sel=np.logical_and(dist>=0,dist<100)
                    vely=vely[sel]
                    dist=dist[sel]
                else: # wells in the lower channel
                    ybottom=0
                    dist=(y-y[-1]+ybottom)*pixelsize
                    sel=np.logical_and(dist>=0,dist<100)
                    vely=-vely[sel]
                    dist=dist[sel]                

                self.vys.append((dist,vely))
                p3+=1    
            p2+=1
```

```python
# datasets 2-6 have different order of wells than dataset 1
# this sorts this out:
orders=[[0,1,2,3,4,5],[0,5,1,2,3,4],[0,5,1,2,3,4],[0,5,1,2,3,4],[0,5,1,2,3,4],[0,5,1,2,3,4]] 
vfs=[VelocityField(currdir+"\\data_Fig_1\\experiment_1\\revision_velocity_profiles_2023_11_14_velocities_"
                   +str(i+1)+".npb",False,orders[i]) for i in range(0,6)]
```

# Fig. S5B

```python
params = {'font.size': 16}
plt.rcParams.update(params)
fig = plt.figure(figsize=(12,3),constrained_layout=True)
fig.patch.set_facecolor('w')
ax_array = fig.subplots(1, 6, squeeze=False)

for pi in range(0,6):
    for ni in range(0,4):
        ax_array[0][pi].plot(vfs[0].vys[4*pi+ni][0],vfs[0].vys[4*pi+ni][1],c='red')
    xx,yy,_=bin_xy(vfs[0].vys[4*pi:4*pi+4])
    imax=np.argmax(yy) # find the point where the growth rate is maximized, and do not go beyond this point
    coeffs=np.polyfit(xx[2:imax],yy[2:imax],deg=1)
    ax_array[0][pi].plot([0,100],[coeffs[1],coeffs[0]*100+coeffs[1]],c='blue')
    ax_array[0][pi].set_title("P="+str(4*origpos[pi])+"-"+str(4*origpos[pi]+3))
    ax_array[0][pi].set_xticks([0,50,100])
    ax_array[0][pi].set_ylim(-5,50)

fig.savefig(currdir+"\\data_Fig_1\\experiment_1\\velocities_in_6_wells.png")
plt.show()
```

# Fig. S5D

```python
params = {'font.size': 16}
plt.rcParams.update(params)
fig = plt.figure(figsize=(12,3),constrained_layout=True)
fig.patch.set_facecolor('w')
ax_array = fig.subplots(1, 6, squeeze=False)

for pi in range(0,6):
    gr_rates=[]
    for ti in range(0,6):
        xx,yy,_=bin_xy(vfs[ti].vys[4*pi:4*pi+4])
        imax=np.argmax(yy) # find the point where the growth rate is maximized, and do not go beyond this point
        coeffs=np.polyfit(xx[2:imax],yy[2:imax],deg=1)
        gr_rates.append(coeffs[0])
        #print(coeffs[0])
    ax_array[0][pi].scatter(times_from_t0,gr_rates,c=[0,1,2,3,4,5])
    ax_array[0][pi].set_ylim(0,1)
fig.savefig(currdir+"\\data_Fig_1\\experiment_1\\growth_rates_in_6_wells.png")
plt.show()
```
