```python
# this function is used for plotting data averaged over time
def plot_av(data,qx=1,out_data=False):
    tmp=np.array(data)
    av=np.mean(tmp[:,:,1],axis=0)
    nsam=len(tmp[:,0,0])
    std=np.std(tmp[:,:,1],axis=0)/np.sqrt(nsam)
    if (out_data==True):
        return tmp[0,:,0]*qx,av,std
    else:
        plt.errorbar(tmp[0,:,0]*qx,av,std)
```

```python
# import data generated by Python script "count_sectors_2chann_v2_5.py"
file=open(currdir+'\\data_Fig_4\\sectors_2023_05_17_rev_fig4_v2_5.csv')
print(file.readline())
secs = [np.array(list(map(int, line.split(',')))) for line in file]
file.close()
secsizes=[s[9:] for s in secs]
secs=np.array([s[0:9] for s in secs])
print(secs[-1])
print(secsizes[-1])
```

```python
# time points for the start/stop of different parts of the protocol
ti_before_RIF=43-1
ti_after_RIF05=43+50-1
ti_after_RIF1=43+50+45-1
ti_at_end=43+50+45+48-1
```

```python
# helper function for getting the average number of sectors vs time
def avsecs(wells,secs):
    tmp=[]
    for s in wells:
        pi=s[0]
        ni=s[1]
        sel=np.logical_and(secs[:,0]==pi,secs[:,3]==ni)
        if (sel.any()==True): tmp.append(secs[sel][:,[1,8]])
    return tmp

def avsecR(wells,secs):
    tmp=[]
    for s in wells:
        pi=s[0]
        ni=s[1]
        sel=np.logical_and(secs[:,0]==pi,secs[:,3]==ni)
        if (sel.any()==True): tmp.append([[x,1.*r/(r+s+1e-6)] for x,s,r in secs[sel][:,[1,4,5]]])
    return tmp

def avsecS(wells,secs):
    tmp=[]
    for s in wells:
        pi=s[0]
        ni=s[1]
        sel=np.logical_and(secs[:,0]==pi,secs[:,3]==ni)
        if (sel.any()==True): tmp.append([[x,1.*s/(r+s+1e-6)] for x,s,r in secs[sel][:,[1,4,5]]])
    return tmp
```

### We first find the mean fractions of green cells before/after RIF

```python
def avsecST(ti,wells,secs):
    #pi,ti,zi,ni,secs,xpos
    tmp=[]
    for s in wells:
        pi=s[0]
        ni=s[1]
        sel=np.logical_and(np.logical_and(secs[:,0]==pi,secs[:,3]==ni),secs[:,1]==ti)
        if (sel.any()==True): tmp.append([1.*s/(r+s) for s,r in secs[sel][:,[4,5]]])
    return np.concatenate(tmp)

def avsector(wells):
    tmp=avsecST(ti_before_RIF,wells,secs)
    return np.mean(tmp), np.std(tmp)/np.sqrt(len(tmp))

data1=np.array([avsector(flat),avsector(T100A10),avsector(T100A20),avsector(T50A10),avsector(T50A20),avsector(T20A10),
               avsector(T20A20),avsector(T10A10),avsector(T10A20)])
print(data1)
```

## Automated counting of sectors

```python
def avsecsT(ti,wells,secs):
    #pi,ti,zi,ni,secs,xpos
    tmp=[]
    for s in wells:
        pi=s[0]
        ni=s[1]
        sel=np.logical_and(np.logical_and(secs[:,0]==pi,secs[:,3]==ni),secs[:,1]==ti)
        if (sel.any()==True): tmp.append(secs[sel][:,7])  # green (sensitive) sectors
    return tmp

def avsector(wells):
    tmp=avsecsT(41,wells,secs)
    return np.mean(tmp), np.std(tmp)/np.sqrt(len(tmp))

data1=np.array([avsector(flat),avsector(T100A10),avsector(T100A20),avsector(T50A10),avsector(T50A20),avsector(T20A10),
               avsector(T20A20),avsector(T10A10),avsector(T10A20)])

def avsector(wells):
    tmp=avsecsT(182,wells,secs)
    return np.mean(tmp), np.std(tmp)/np.sqrt(len(tmp))


data2=np.array([avsector(flat),avsector(T100A10),avsector(T100A20),avsector(T50A10),avsector(T50A20),avsector(T20A10),
               avsector(T20A20),avsector(T10A10),avsector(T10A20)])
plt.rcParams["figure.figsize"] = (8,4)
fig,ax=plt.subplots()
fig.patch.set_facecolor('w')
x_pos=[0,1,2,3,4,5,6,7,8]
ax.set_xticks(x_pos)
ax.set_xticklabels(["flat","T=100\nA=5.1","T=100\nA=9.5","T=50\nA=4.6","T=50\nA=8.7","T=20\nA=3.3","T=20\nA=5.1","T=10\nA=1.3","T=10\nA=1.7"])
ax.set_ylabel('average no. of sensitive sectors')
plt.figure(facecolor='w')
ax.bar(x_pos,data1[:,0],yerr=data1[:,1],align='center', alpha=0.5, ecolor='black', capsize=10, label="before RIF")
ax.bar(x_pos,data2[:,0],yerr=data2[:,1],align='center', alpha=0.5, ecolor='black', capsize=10, label="after high RIF + LB")
ax.legend()
fig.savefig(currdir+"\\data_Fig_4\\aver_sectors_before_after_automated.png")
plt.show()
```
